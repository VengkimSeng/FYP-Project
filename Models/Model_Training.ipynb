# Model Training TF-IDF

````python
# Import libraries
import os
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize, LabelBinarizer
import joblib
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Define paths
PROCESSED_TEXTS_DIR = r'E:\Paragon\PIU Year 4\MIS 401 - FYP I\FYP_CODE\FYP-KhmerClassificationModel/preprocessed_articles'
METADATA_PATH = r'E:\Paragon\PIU Year 4\MIS 401 - FYP I\FYP_CODE\FYP-KhmerClassificationModel/metadata.csv'
MODELS_DIR = r'E:\Paragon\PIU Year 4\MIS 401 - FYP I\FYP_CODE\FYP-KhmerClassificationModel/models'

# Create models directory if it doesn't exist
os.makedirs(MODELS_DIR, exist_ok=True)

# Load metadata
print("Loading metadata...")
metadata = pd.read_csv(METADATA_PATH)
print(f"Metadata loaded with {len(metadata)} entries")

# Display metadata sample
metadata.head()

# Get category distribution
category_counts = metadata['category'].value_counts()
print("\nCategory distribution:")
print(category_counts)

# Plot category distribution
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title('Number of Articles per Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Custom analyzer for Khmer text
def khmer_word_analyzer(text):
    # Since the text is already segmented with proper spacing,
    # we can split by spaces and keep words in Khmer Unicode range
    words = []
    for word in text.split():
        # Check if the word contains Khmer characters (Unicode range \u1780-\u17FF)
        if any(('\u1780' <= char <= '\u17FF') for char in word):
            # Remove any non-Khmer characters like punctuation
            khmer_word = ''.join(char for char in word if '\u1780' <= char <= '\u17FF' or char == '្')
            if len(khmer_word) >= 2:  # Only keep words with at least 2 characters
                words.append(khmer_word)
    return words

# Function to load preprocessed texts
def load_texts_and_labels():
    print("Loading texts and labels...")
    texts = []
    labels = []
    doc_ids = []
    
    total = len(metadata)
    for i, (_, row) in enumerate(metadata.iterrows()):
        # Print progress every 500 documents
        if i % 500 == 0 or i == total - 1:
            print(f"Progress: {i+1}/{total} documents loaded ({(i+1)/total*100:.1f}%)")
            
        doc_id = row['docId']
        category = row['category']
        file_path = os.path.join(PROCESSED_TEXTS_DIR, f"{doc_id}.txt")
        
        if os.path.exists(file_path):
            with open(file_path, 'r', encoding='utf-8') as f:
                text = f.read()
            texts.append(text)
            labels.append(category)
            doc_ids.append(doc_id)
    
    return texts, labels, doc_ids

# Load texts and labels
texts, labels, doc_ids = load_texts_and_labels()
print(f"Loaded {len(texts)} documents with {len(set(labels))} unique categories")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)
print(f"Training set: {len(X_train)} documents")
print(f"Test set: {len(X_test)} documents")

# Create TF-IDF features
print("Creating TF-IDF features...")
tfidf_vectorizer = TfidfVectorizer(
    analyzer=khmer_word_analyzer,
    max_features=None,  # No limit on features
    min_df=5,           # Term must appear in at least 5 documents
    max_df=0.8,         # Term must appear in at most 80% of documents
    ngram_range=(1, 2)  # Include both single words and bigrams
)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print(f"TF-IDF features shape: {X_train_tfidf.shape}")
print(f"Number of features: {X_train_tfidf.shape[1]}")

# Save the vectorizer
joblib.dump(tfidf_vectorizer, os.path.join(MODELS_DIR, 'tfidf_vectorizer.joblib'))
print("TF-IDF vectorizer saved")

# Function to train and evaluate a model with cross-validation
def train_and_evaluate_model(model_name, model, X_train, y_train, X_test, y_test):
    print(f"\n{'='*50}")
    print(f"Training and evaluating {model_name}...")
    start_time = time.time()
    
    # Perform cross-validation to assess model performance
    print(f"Performing 5-fold cross-validation...")
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"Cross-validation accuracies: {cv_scores}")
    print(f"Mean CV accuracy: {cv_scores.mean():.4f} (±{cv_scores.std():.4f})")
    
    # Train the model on the full training set
    print(f"Training final model on complete training set...")
    model.fit(X_train, y_train)
    
    # Calculate training accuracy
    train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(y_train, train_pred)
    print(f"Training accuracy: {train_accuracy:.4f}")
    
    # Calculate training time
    training_time = time.time() - start_time
    print(f"Training completed in {training_time:.2f} seconds")
    
    # Save the model
    joblib.dump(model, os.path.join(MODELS_DIR, f'{model_name.lower().replace(" ", "_")}.joblib'))
    print(f"{model_name} model saved")
    
    # Evaluate on test set
    print(f"\nEvaluating {model_name} on test set...")
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    print(f"Test Accuracy: {accuracy:.4f}")
    print(f"Test Precision: {precision:.4f}")
    print(f"Test Recall: {recall:.4f}")
    print(f"Test F1 Score: {f1:.4f}")
    
    # Create and plot confusion matrix
    print("\nGenerating confusion matrix...")
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(12, 10))
    categories = sorted(set(labels))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=categories, yticklabels=categories)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(MODELS_DIR, f'{model_name.lower().replace(" ", "_")}_confusion_matrix.png'))
    plt.show()
    
    # Create learning curve plot (training size vs accuracy)
    print("\nGenerating learning curve...")
    train_sizes = np.linspace(0.1, 1.0, 5)
    try:
        from sklearn.model_selection import learning_curve
        train_sizes, train_scores, val_scores = learning_curve(
            model, X_train, y_train, train_sizes=train_sizes, cv=5, scoring='accuracy'
        )
        
        train_mean = np.mean(train_scores, axis=1)
        train_std = np.std(train_scores, axis=1)
        val_mean = np.mean(val_scores, axis=1)
        val_std = np.std(val_scores, axis=1)
        
        plt.figure(figsize=(10, 6))
        plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training score')
        plt.plot(train_sizes, val_mean, 'o-', color='g', label='Validation score')
        plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
        plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.1, color='g')
        plt.title(f'Learning Curve - {model_name}')
        plt.xlabel('Training Set Size')
        plt.ylabel('Accuracy Score')
        plt.legend(loc='best')
        plt.grid(True)
        plt.savefig(os.path.join(MODELS_DIR, f'{model_name.lower().replace(" ", "_")}_learning_curve.png'))
        plt.show()
    except Exception as e:
        print(f"Warning: Could not generate learning curve. Error: {e}")
    
    # Generate classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    # Create ROC curves for multiclass (One-vs-Rest)
    try:
        print("\nGenerating ROC curves (one-vs-rest)...")
        # Binarize labels for ROC curve calculation
        lb = LabelBinarizer()
        lb.fit(labels)
        y_test_bin = lb.transform(y_test)
        
        # Get decision function or probability estimates if available
        if hasattr(model, "decision_function"):
            y_score = model.decision_function(X_test)
        else:
            # For models like MultinomialNB that don't have decision_function
            y_score = model.predict_proba(X_test)
        
        # Calculate ROC curve and ROC area for each class
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        
        n_classes = len(set(labels))
        
        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        
        # Plot ROC curves
        plt.figure(figsize=(12, 8))
        
        # Only plot a subset of categories if there are many
        categories_to_plot = set(labels)
        if len(categories_to_plot) > 10:
            # Take only the top 10 categories by frequency
            categories_to_plot = category_counts.index[:10]
        
        for i, category in enumerate(lb.classes_):
            if category in categories_to_plot:
                plt.plot(fpr[i], tpr[i], lw=2, 
                         label=f'{category} (AUC = {roc_auc[i]:.2f})')
        
        plt.plot([0, 1], [0, 1], 'k--', lw=2)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curves (One-vs-Rest) - {model_name}')
        plt.legend(loc="lower right")
        plt.savefig(os.path.join(MODELS_DIR, f'{model_name.lower().replace(" ", "_")}_roc_curves.png'))
        plt.show()
    except Exception as e:
        print(f"Warning: Could not generate ROC curves. Error: {e}")
    
    return {
        'model_name': model_name,
        'cv_scores': cv_scores,
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'training_time': training_time,
        'train_accuracy': train_accuracy
    }

print("\n" + "="*80)
print("STEP 1: Identifying top features for each category")
print("="*80)

# Initialize a dictionary to store top features for each category
category_top_features = {}
top_n_features = 5000  # Maximum number of top features per category

# Get unique categories
categories = sorted(set(y_train))
print(f"Found {len(categories)} categories: {categories}")

# For each category, find the top features
for category in categories:
    print(f"\nProcessing category: {category}")
    
    # Get indices of documents in this category
    category_indices = [i for i, label in enumerate(y_train) if label == category]
    print(f"Found {len(category_indices)} documents in category '{category}'")
    
    # Extract the TF-IDF matrix for only this category's documents
    category_tfidf = X_train_tfidf[category_indices]
    
    # Calculate the average TF-IDF score for each feature across category documents
    feature_scores = np.array(category_tfidf.mean(axis=0)).flatten()
    
    # Get feature names
    feature_names = tfidf_vectorizer.get_feature_names_out()
    
    # Create a DataFrame with features and scores
    feature_df = pd.DataFrame({'feature': feature_names, 'score': feature_scores})
    
    # Sort by score in descending order
    feature_df = feature_df.sort_values('score', ascending=False)
    
    # Get top N features (or less if there are fewer features)
    top_n = min(top_n_features, len(feature_df))
    top_features = feature_df.head(top_n)
    
    # Store top features for this category
    category_top_features[category] = set(top_features['feature'])
    
    print(f"Extracted top {top_n} features for category '{category}'")
    
    # Optional: save these features to file
    output_path = os.path.join(MODELS_DIR, f"{category}_top_features.txt")
    with open(output_path, 'w', encoding='utf-8') as f:
        for _, row in top_features.iterrows():
            f.write(f"{row['feature']}\t{row['score']:.6f}\n")
    print(f"Saved to: {output_path}")

print("\n" + "="*80)
print("STEP 2: Finding common features across all categories")
print("="*80)

# Find common features across all categories
common_features = set()
if len(category_top_features) > 0:
    # Get the intersection of all sets of top features
    common_features = set.intersection(*category_top_features.values())
    print(f"Number of common features across all categories: {len(common_features)}")
    
    # Display some examples of common features (if any)
    if common_features:
        print("\nSample of common features (up to 20):")
        for feature in list(common_features)[:20]:
            print(f"- {feature}")
        
        # Save common features to file
        common_features_path = os.path.join(MODELS_DIR, "common_features.txt")
        with open(common_features_path, 'w', encoding='utf-8') as f:
            f.write(f"Total common features found: {len(common_features)}\n\n")
            for feature in sorted(common_features):
                f.write(f"{feature}\n")
        print(f"Saved {len(common_features)} common features to: {common_features_path}")
else:
    print("No categories found to process")

print("\n" + "="*80)
print("STEP 3: Removing common features and creating enhanced feature sets")
print("="*80)

# Create enhanced feature sets by removing common features
enhanced_category_features = {}
for category, features in category_top_features.items():
    # Remove common features
    enhanced_features = features - common_features
    enhanced_category_features[category] = enhanced_features
    
    print(f"\nCategory: {category}")
    print(f"Original features: {len(features)}")
    print(f"Enhanced features (after removing {len(common_features)} common features): {len(enhanced_features)}")
    
    # Save enhanced features to file
    output_path = os.path.join(MODELS_DIR, f"{category}_enhanced_features.txt")
    
    # Read the original file to get scores
    original_path = os.path.join(MODELS_DIR, f"{category}_top_features.txt")
    feature_scores = {}
    with open(original_path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split('\t')
            if len(parts) == 2:
                feature, score = parts
                feature_scores[feature] = float(score)
    
    # Write enhanced features with scores
    with open(output_path, 'w', encoding='utf-8') as f:
        for feature in sorted(enhanced_features, key=lambda x: feature_scores.get(x, 0), reverse=True):
            score = feature_scores.get(feature, 0)
            f.write(f"{feature}\t{score:.6f}\n")
    
    print(f"Saved enhanced features to: {output_path}")

print("\n" + "="*80)
print("STEP 4: Creating new TF-IDF matrix with enhanced features")
print("="*80)

# Extract all enhanced features (union of all category-specific enhanced features)
all_enhanced_features = set()
for features in enhanced_category_features.values():
    all_enhanced_features.update(features)
print(f"Total unique enhanced features across all categories: {len(all_enhanced_features)}")

# Create a new vectorizer that only includes the enhanced features
enhanced_vectorizer = TfidfVectorizer(
    analyzer=khmer_word_analyzer,
    vocabulary=all_enhanced_features,  # Only use the enhanced features
    ngram_range=(1, 2)
)

# Transform the data using the enhanced vectorizer
X_train_enhanced = enhanced_vectorizer.fit_transform(X_train)
X_test_enhanced = enhanced_vectorizer.transform(X_test)

print(f"Original TF-IDF feature matrix shape: {X_train_tfidf.shape}")
print(f"Enhanced TF-IDF feature matrix shape: {X_train_enhanced.shape}")

# Save the enhanced vectorizer
joblib.dump(enhanced_vectorizer, os.path.join(MODELS_DIR, 'enhanced_tfidf_vectorizer.joblib'))
print("Enhanced TF-IDF vectorizer saved")

print("\n" + "="*80)
print("Training models on original TF-IDF features")
print("="*80)

# Train SVM with specified parameters on original features
svm_model_original = SVC(
    C=1, 
    kernel='rbf',
    gamma=0.01,
    decision_function_shape='ovr',
    random_state=42
)
svm_results_original = train_and_evaluate_model(
    "SVM (Original Features)", 
    svm_model_original, 
    X_train_tfidf, y_train, 
    X_test_tfidf, y_test
)

# Train Multinomial Naive Bayes on original features
mnb_model_original = MultinomialNB()
mnb_results_original = train_and_evaluate_model(
    "MNB (Original Features)", 
    mnb_model_original, 
    X_train_tfidf, y_train, 
    X_test_tfidf, y_test
)

print("\n" + "="*80)
print("Training models on enhanced features (common features removed)")
print("="*80)

# Train SVM with specified parameters on enhanced features
svm_model_enhanced = SVC(
    C=1, 
    kernel='rbf',
    gamma=0.01,
    decision_function_shape='ovr',
    random_state=42
)
svm_results_enhanced = train_and_evaluate_model(
    "SVM (Enhanced Features)", 
    svm_model_enhanced, 
    X_train_enhanced, y_train, 
    X_test_enhanced, y_test
)

# Train Multinomial Naive Bayes on enhanced features
mnb_model_enhanced = MultinomialNB()
mnb_results_enhanced = train_and_evaluate_model(
    "MNB (Enhanced Features)", 
    mnb_model_enhanced, 
    X_train_enhanced, y_train, 
    X_test_enhanced, y_test
)

# Compare all models together
results = [
    svm_results_original,
    svm_results_enhanced,
    mnb_results_original,
    mnb_results_enhanced
]
model_comparison = pd.DataFrame(results)

print("\nModel Comparison (Original vs Enhanced Features):")
comparison_cols = ['model_name', 'cv_mean', 'accuracy', 'precision', 'recall', 'f1', 'training_time']
print(model_comparison[comparison_cols])

# Plot model comparison
metrics = ['cv_mean', 'accuracy', 'precision', 'recall', 'f1']
metric_labels = ['CV Accuracy', 'Test Accuracy', 'Precision', 'Recall', 'F1 Score']
model_names = [result['model_name'] for result in results]

plt.figure(figsize=(15, 8))
x = np.arange(len(metrics))
width = 0.2  # Width of bars, adjusted for 4 models

colors = ['royalblue', 'lightblue', 'firebrick', 'salmon']
for i, result in enumerate(results):
    values = [result[metric] for metric in metrics]
    plt.bar(x + (i-1.5)*width, values, width, label=result['model_name'], color=colors[i])

plt.xlabel('Metric')
plt.ylabel('Score')
plt.title('Model Performance Comparison: Original vs Enhanced Features')
plt.xticks(x, metric_labels)
plt.legend(loc='lower left', bbox_to_anchor=(0, -0.15), ncol=2)
plt.ylim(0.7, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(rect=[0, 0.05, 1, 1])  # Adjust layout to make room for legend
plt.savefig(os.path.join(MODELS_DIR, 'original_vs_enhanced_comparison.png'))
plt.show()

# Calculate improvement percentages
improvement_svm = (svm_results_enhanced['accuracy'] - svm_results_original['accuracy']) / svm_results_original['accuracy'] * 100
improvement_mnb = (mnb_results_enhanced['accuracy'] - mnb_results_original['accuracy']) / mnb_results_original['accuracy'] * 100

print(f"\nFeature Enhancement Impact:")
print(f"SVM accuracy change: {improvement_svm:.2f}%")
print(f"MNB accuracy change: {improvement_mnb:.2f}%")

# Determine best overall model
all_accuracies = [
    svm_results_original['accuracy'],
    svm_results_enhanced['accuracy'],
    mnb_results_original['accuracy'],
    mnb_results_enhanced['accuracy']
]
best_model_idx = np.argmax(all_accuracies)
best_model_name = model_names[best_model_idx]

print(f"\nBest performing model overall: {best_model_name} with accuracy: {all_accuracies[best_model_idx]:.4f}")

# Save best model
best_model = None
if best_model_idx == 0:
    best_model = svm_model_original
    best_vectorizer = tfidf_vectorizer
elif best_model_idx == 1:
    best_model = svm_model_enhanced
    best_vectorizer = enhanced_vectorizer
elif best_model_idx == 2:
    best_model = mnb_model_original
    best_vectorizer = tfidf_vectorizer
else:
    best_model = mnb_model_enhanced
    best_vectorizer = enhanced_vectorizer

# Save as the final model
joblib.dump(best_model, os.path.join(MODELS_DIR, 'best_model.joblib'))
joblib.dump(best_vectorizer, os.path.join(MODELS_DIR, 'best_vectorizer.joblib'))
print(f"Best model and vectorizer saved for production use")

# Function for error analysis
def analyze_errors(model, X_test, y_test, categories, top_n=10):
    print(f"\n{'='*50}")
    print(f"Error Analysis for {model.__class__.__name__}")
    
    # Predict on test set
    y_pred = model.predict(X_test)
    
    # Identify misclassified instances
    misclassified_indices = np.where(np.array(y_test) != np.array(y_pred))[0]
    
    if len(misclassified_indices) == 0:
        print("No misclassified instances found.")
        return
    
    print(f"Found {len(misclassified_indices)} misclassified instances out of {len(y_test)} test samples.")
    
    # Group errors by true category
    errors_by_category = {}
    for idx in misclassified_indices:
        true_category = y_test[idx]
        pred_category = y_pred[idx]
        
        if true_category not in errors_by_category:
            errors_by_category[true_category] = []
        
        errors_by_category[true_category].append(pred_category)
    
    # Calculate error rate by category
    error_rates = {}
    for category in categories:
        # Count samples from this category in the test set
        category_count = sum(1 for y in y_test if y == category)
        if category_count == 0:
            continue
        
        # Count misclassifications for this category
        misclassified_count = len(errors_by_category.get(category, []))
        error_rate = misclassified_count / category_count
        error_rates[category] = error_rate
    
    # Display top N categories with highest error rates
    print("\nCategories with Highest Error Rates:")
    sorted_categories = sorted(error_rates.items(), key=lambda x: x[1], reverse=True)
    for category, error_rate in sorted_categories[:top_n]:
        correct_count = sum(1 for y in y_test if y == category) - len(errors_by_category.get(category, []))
        total_count = sum(1 for y in y_test if y == category)
        print(f"{category}: {error_rate:.2%} error rate ({correct_count}/{total_count} correctly classified)")
    
    # Common confusion pairs
    confusion_pairs = {}
    for true_category, errors in errors_by_category.items():
        for pred_category in errors:
            pair = (true_category, pred_category)
            confusion_pairs[pair] = confusion_pairs.get(pair, 0) + 1
    
    print("\nMost Common Confusion Pairs (True -> Predicted):")
    sorted_pairs = sorted(confusion_pairs.items(), key=lambda x: x[1], reverse=True)
    for (true_cat, pred_cat), count in sorted_pairs[:top_n]:
        print(f"{true_cat} -> {pred_cat}: {count} instances")

# Perform error analysis on the best model
print("\n" + "="*80)
print("Error Analysis for Best Model")
print("="*80)

best_model_data = None
if best_model_idx == 0:
    best_model_data = (svm_model_original, X_test_tfidf)
elif best_model_idx == 1:
    best_model_data = (svm_model_enhanced, X_test_enhanced)
elif best_model_idx == 2:
    best_model_data = (mnb_model_original, X_test_tfidf)
else:
    best_model_data = (mnb_model_enhanced, X_test_enhanced)

analyze_errors(best_model_data[0], best_model_data[1], y_test, categories)

print("\n" + "="*80)
print("Training and Evaluation Complete!")
print("="*80)
print(f"All models and vectorizers saved in: {MODELS_DIR}")
print(f"Best model: {best_model_name}")
print("\nUse the following code to load and use the best model for prediction:")
print("```python")
print("import joblib")
print("# Load best model and vectorizer")
print("model = joblib.load('path/to/best_model.joblib')")
print("vectorizer = joblib.load('path/to/best_vectorizer.joblib')")
print("# For prediction:")
print("def khmer_word_analyzer(text):")
print("    # Custom analyzer implementation (same as in training)")
print("    pass")
print("# Transform new text")
print("X_new_tfidf = vectorizer.transform([new_text])")
print("# Predict")
print("prediction = model.predict(X_new_tfidf)")
print("```")
````
